```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
raw_data_path = r'C:\Users\Antoine Dupont\Pictures\House-Prediction\data\raw\train.csv'
df = pd.read_csv(raw_data_path)
print("Raw data shape:", df.shape)
df.head()
```

```python
df = df.drop(columns=['Id'])
print("Removed Id column. Shape:", df.shape)
```

```python
target = df['SalePrice']
print("Target separated.")
```

```python
selected_numerics = [
    'OverallQual',
    'GrLivArea',
    'GarageCars',
    'GarageArea'
]

selected_categoricals = [
    'ExterQual',
    'BsmtQual',
    'KitchenQual',
    'GarageFinish',
    'GarageType'
]

final_features = selected_numerics + selected_categoricals
final_features.append('SalePrice')
```

```python
df = df[final_features]
print("Subset shape:", df.shape)
df.head()
```

```python
for col in selected_numerics:
    df[col] = df[col].fillna(df[col].median())

for col in selected_categoricals:
    df[col] = df[col].fillna('Missing')

print("Missing values handled.")

```

```python
le = LabelEncoder()

for col in selected_categoricals:
    df[col] = le.fit_transform(df[col])

print("Categorical encoding complete.")
```

```python
print(df.info())
df.head()
```

```python
processed_data_path = '../../data/processed/processed_train.csv'
df.to_csv(processed_data_path, index=False)
print(f"Saved processed data to {processed_data_path}")
```
